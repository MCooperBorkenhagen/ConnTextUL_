```python
import os
import pandas as pd
import chardet
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime
formatted_date=datetime.today().strftime("%m%d%y")
base_directory = r"C:\Users\razza\ConnTextUL_data\data\vocabulary\tasa.csv"
tasa = pd.read_csv(base_directory)
tasa['word'] = tasa['word'].str.lower()
tasa = tasa.dropna(subset=['word'])
```

```python
# Define a function to filter by levels
def filter_by_level(df, max_level):
    return df[df['level'].between(1, max_level)]
```

```python
max_level_to_include = 3

# Apply the filter_by_level function
filtered_tasa = filter_by_level(tasa, max_level_to_include)

# Step 4: Filter out non-alphabetical words
filtered_tasa = filtered_tasa[filtered_tasa['word'].str.match(r'^[a-z]+$')]

# Step 5: Group by the "word" column and sum the "frequency" column
word_frequency_tasa = filtered_tasa.groupby('word')['frequency'].sum().reset_index()

# Step 6: Rename the columns
word_frequency_tasa.columns = ['unique_words', 'frequency']

# Display the new DataFrame
print(word_frequency_tasa)
```

```python
base_directory = r"C:\Users\razza\ConnTextUL_data\data\vocabulary\tidycorpus.csv"
data = []
tidy = pd.read_csv(base_directory)
tidy['token'] = tidy['token'].str.lower()
tidy['token'] = tidy['token'].str.replace(r'[^a-zA-Z\s]', '', regex=True)
tidy = tidy.dropna(subset=['token'])
tidy
```

```python
# Count the frequency of each word
word_frequency_tidy = tidy['token'].value_counts().reset_index()
# Rename columns to 'unique_words' and 'frequency'
word_frequency_tidy.columns = ['unique_words', 'frequency']
word_frequency_tidy = word_frequency_tidy[word_frequency_tidy['unique_words'].str.strip().astype(bool)]
# Display the new DataFrame
print(word_frequency_tidy)
```

```python
word_frequency_combined = pd.merge(word_frequency_tasa, word_frequency_tidy, on='unique_words', how='left')
word_frequency_combined['frequency_y'].fillna(0, inplace=True)
# Create a new column 'combined_frequency' by adding the frequencies, considering your condition
word_frequency_combined['combined_frequency'] = word_frequency_combined['frequency_x'] + word_frequency_combined['frequency_y']
# Rename the columns 'unique_words' to 'words' and 'frequency' to 'word_frequency'
word_frequency_combined = word_frequency_combined.rename(columns={'frequency_x': 'frequency_tasa', 'frequency_y': 'frequency_tidy'})
word_frequency_combined
```

```python
# Calculate the total word count
total_word_count = word_frequency_combined['combined_frequency'].sum()
# Calculate the average per million value
word_frequency_combined['average_per_million'] = (word_frequency_combined['combined_frequency'] / total_word_count) * 1e6

# Calculate the total "average_per_million" value
total_average_per_million = word_frequency_combined['average_per_million'].sum()

# Calculate the sampling probability based on "average_per_million"
word_frequency_combined['sampling_probability_per_million'] = word_frequency_combined['average_per_million'] / total_average_per_million
# Calculate the sampling probability 
word_frequency_combined['sampling_probability'] = word_frequency_combined['combined_frequency'] / total_word_count
# Display the new DataFrame with the sampling probability
print(word_frequency_combined)
```
